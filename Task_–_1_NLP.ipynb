**1. What is the difference between "Love" and "love" in NLP?**

At first glance, "Love" and "love" seem identical, but in NLP they are treated as two completely different tokens because of case sensitivity. This means that a model might interpret them differently even though they carry the same meaning in most contexts.

For example, "Love" could appear at the beginning of a sentence or be used as a proper noun (like a name), while "love" is generally used as a verb or noun in everyday language. If we don’t normalize the text by converting everything to lowercase, the model ends up treating them as separate words, which unnecessarily increases the vocabulary size.

This is why lowercasing is an important preprocessing step — it helps in reducing redundancy and ensures that similar words are grouped together, improving model efficiency and learning.


**2. What happens if stopwords are not removed?**

Stopwords are very common words like “is”, “the”, “and”, “in”, etc. If we don’t remove them, they can introduce a lot of noise into the dataset.

Since these words appear very frequently but don’t carry strong meaning on their own, they can dominate the dataset and make it harder for the model to focus on important words. This can lead to:

Increased computational cost
Larger vocabulary size
Reduced model performance

However, it’s important to note that removing stopwords is not always the best choice. In some cases, these words play a crucial role in understanding the actual meaning of a sentence.


**3. Provide two real-world scenarios where removing stopwords can be harmful**

There are situations where removing stopwords can completely change the meaning of a sentence:

1. Sentiment Analysis
In sentiment-based tasks, words like “not” are extremely important.
Example:

Original: "I am not happy"
After removing stopwords: "happy"

This completely reverses the meaning of the sentence, which can lead to incorrect predictions.

2. Question Answering Systems / Chatbots
In systems that rely on understanding user intent, stopwords are essential.
Example:

Original: "What is your name?"
After removing stopwords: "name"

This removes the structure of the question and makes it harder for the system to understand what the user is asking.


**4. Explain the difference between stemming and lemmatization**

Both stemming and lemmatization are techniques used to reduce words to their base or root form, but they work in very different ways.

Stemming is a simple, rule-based approach. It removes suffixes from words without considering their actual meaning. Because of this, it can sometimes produce incorrect or incomplete words.
Example:

“running” → “runn”

“studies” → “studi”

While stemming is fast and computationally efficient, its outputs are not always meaningful.

Lemmatization, on the other hand, is more advanced and context-aware. It uses vocabulary and linguistic rules to convert words into their correct base form (called a lemma).
Example:

“running” → “run”

“better” → “good”

Although lemmatization is slower than stemming, it produces more accurate and meaningful results, which makes it more suitable for real-world NLP applications.

# Task 2

```python
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove emojis & special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

# Task 3

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sent = preprocess_text(sentence)
    results.append((sentence, tokens, clean_sent))

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sent)
```

# Task 4

```python
analytics = []

for sentence, tokens, clean_sent in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    analytics.append({
        "sentence": sentence,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_length": avg_length
    })

# Print analytics
for item in analytics:
    print("\nSentence:", item["sentence"])
    print("Total Tokens:", item["total_tokens"])
    print("Unique Tokens:", item["unique_tokens"])
    print("Avg Token Length:", round(item["avg_length"], 2))
```

# Task 5

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nLeast 5 Frequent Words:")
print(counter.most_common()[-5:])
```

# Task 6

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sent = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sent)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)
print(output)
```
